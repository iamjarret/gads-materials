```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.cross_validation import train_test_split
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression #http://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.f_regression.html
from sklearn.metrics import mean_squared_error
from sklearn.cross_validation import KFold
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
# ignore DeprecateWarnings by sklearn
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)

# adapted from http://nbviewer.ipython.org/github/sujitpal/statlearning-notebooks/blob/master/src/chapter6.ipynb
```

```python
df = pd.read_csv('/home/vagrant/notebooks/fall_2014_lessons/datasets/Hitters.csv')
# drop NaN values
df.dropna(inplace=True)
# Converting non-numeric input variables to factors.
df["League"] = pd.factorize(df["League"])[0]
df["Division"] = pd.factorize(df["Division"])[0]
df["NewLeague"] = pd.factorize(df["NewLeague"])[0]
predictors = [col for col in df.columns if col != "Salary"]
```

```python
df[predictors].corr()
```

```python
X = df[predictors].values
y = df.Salary.values
```

### Find an optimum value of alpha (lambda) for the Ridge regressor using cross-validation

```python
np.linspace(0,1,10)
```

```python

cv_errors = []

alphas = np.linspace(0,1, 100) #[0.0, 1e-9, 1e-5, 1]
kfold = KFold(X.shape[0], n_folds=10)
for alpha in alphas:
    rmses = []
    for train, test in kfold:
        Xtrain, ytrain, Xtest, ytest = X[train], y[train], X[test], y[test]
        reg = Ridge(alpha=alpha)
        reg.fit(Xtrain, ytrain)
        ypred = reg.predict(Xtest)
        rmses.append(np.sqrt(mean_squared_error(ytest, ypred)))
    cv_errors.append(np.mean(rmses))
fig, ax = plt.subplots(figsize=(8, 6))
# prevent matplotlib from converting to scientific notation
y_formatter = matplotlib.ticker.ScalarFormatter(useOffset=False)
ax.yaxis.set_major_formatter(y_formatter)

plt.plot(alphas, cv_errors)
plt.xlabel("alpha")
plt.ylabel("RMSE")
```

### Find an optimum value of alpha (lambda) for the Lasso regressor using cross-validation

```python
cv_errors = []

alphas = np.linspace(0,2, 100) #[0.1 * alpha for alpha in range(1, 12)]
kfold = KFold(X.shape[0], n_folds=10)
for alpha in alphas:
    rmses = []
    for train, test in kfold:
        Xtrain, ytrain, Xtest, ytest = X[train], y[train], X[test], y[test]
        reg = Lasso(alpha=alpha)
        reg.fit(Xtrain, ytrain)
        ypred = reg.predict(Xtest)
        rmses.append(np.sqrt(mean_squared_error(ytest, ypred)))
    cv_errors.append(np.mean(rmses))
plt.plot(alphas, cv_errors)
plt.xlabel("alpha")
plt.ylabel("RMSE")
```
