```python
import pandas
import sklearn
```

```python
africa_data = pandas.read_csv('../africasoil/training.csv')
```

```python
colnames=africa_data.columns.values
```

```python
dependent = ['SOC', 'pH','Ca','P','Sand']
sampleid = ['PIDN']
```

```python
covariates=[]
for x in colnames:
    if x not in dependent+sampleid:
        covariates.append(x)
```

```python
infrared=[]
for x in covariates:
    if x[0]=='m':
        infrared.append(x)
    
```

# Using PCA to Summarize High Dimensional Data

```python
#create new vectors from infrared data with PCA
from sklearn.decomposition import PCA
```

```python
PCA1= PCA(n_components=0.95)
components =PCA1.fit_transform(africa_data[infrared])
```

### PCA iteratively finds the 1-dimensional mapping that maximizes variance

```python
explained_variance = [PCA1.explained_variance_ratio_[0:i].sum() for i in range(6)]
plot([0,1,2,3,4,5],explained_variance)
```

### Each component is a linear transformation of the existing data

```python
plot(PCA1.components_[0,:])
```

### That are orthogonal

```python
from IPython.display import Image
Image(url='http://upload.wikimedia.org/wikipedia/commons/1/15/GaussianScatterPCA.png')
```

### But not necessarily uncorrelated

```python
PCAdict = {'1st PCA Component':PCA1.components_[0,:], 
        '2nd PCA Component':PCA1.components_[1,:],
        '3rd PCA Component':PCA1.components_[2,:],
        '4th PCA Component':PCA1.components_[3,:],
        '5th PCA Component':PCA1.components_[4,:]}
PCADF = pandas.DataFrame.from_dict(PCAdict)
```

```python
PCADF.corr()
```

### PCA is not robust to overfitting

```python
import random
def emptyorlarge(m):
    if random.random()<0.95:
        return 0
    else:
        return m
```

```python
africa_data['High_Var_Infrared']=africa_data[infrared[0]].apply(lambda x: emptyorlarge(10000))
```

```python
africa_data['High_Var_Infrared'].var()
```

```python
newinfrared = infrared + ['High_Var_Infrared']
PCA2= PCA(n_components=5)
components =PCA2.fit_transform(africa_data[newinfrared])
```
