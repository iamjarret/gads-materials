```python
def add_infrared_variables(dataset, infrared=infrared):
    dataset['Infrared01'] = dataset[infrared[500]]
    dataset['Infrared02'] = dataset[infrared[1100:1400]].max(axis=1)-dataset['Infrared01'] #bump
    dataset['Infrared03'] = dataset[infrared[1900:2100]].max(axis=1)-dataset['Infrared01'] #peak1
    #dataset['wave_amplitude'] = dataset[infrared[1980:2750]].max(axis=1)-dataset[infrared[1980:2750]].min(axis=1)
    dataset['Infrared04'] = dataset[infrared[3150:3220]].max(axis=1) #peak2
    dataset['Infrared05'] = dataset[infrared[3235:3300]].max(axis=1) #finaltrough
add_infrared_variables(africa_data)
```

```python
def add_poly_terms(data, covariates):
    polynomialterms=[]
    for h in covariates:
        #add 2nd and 3rd polynomials
        name2 = "%s**2"%h
        polynomialterms.append(name2)
        name3 = "%s**3"%h
        polynomialterms.append(name3)
        data[name2]=data[h]**2
        data[name3]=data[h]**3
    return polynomialterms
```

```python
def add_interact_terms(data, covariates):
    interactionterms=[]
    for h in covariates:
        for m in covariates:
            if h != m:
                name = "%s-%s"%(h,m)
                interactionterms.append(name)
                data[name]=data[h]*africa_soil[m]
    return interactionterms
            
```

```python
def custom_grid_search(data, y, models, num_cv = 5, variables=[], model_metrics=[]):
    results = {h: [] for h in models.keys()}
    models = models.copy()
    for i in range(num_cv):
        in_train = [np.random.random()>0.5 for i in range(data.shape[0])]
        out_train = [not h for h in in_train]
        train =data[in_train]
        test = data[out_train]
        for model in models:
            full_model = models[model] + variables
            ph_model.fit(train[full_model],train[y])
            results[model].append(mean_squared_error(test[y],ph_model.predict(test[full_model])))
    results = {h: max(results[h]) for h in results}
    plt.hist(results.values())
    plt.show()
    variable = sorted(results,key=results.get, reverse=False)[0]
    final_metric = results[variable]
    print variable, final_metric
    if len(variables)>5:
        if final_metric<model_metrics[-1] or len(variables)>10:
            return (variables, model_metrics)
    del models[variable]
    variables.append(variable)
    model_metrics.append(final_metric)
    return custom_grid_search(data,y, models=models, num_cv=num_cv, variables=variables, model_metrics=model_metrics)
```
