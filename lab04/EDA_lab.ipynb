```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./nyt1.csv")
df = df[df.Signed_In==1]
df = df[df.Impressions>0]
```

```python
df[df.Impressions==1].count()
```

```python
df.describe()
```

```python
# Create a new variable "age_group" that puts users into categories: ["<18", "18-24", "25-34", "35-44", "45-54", "55-64", "65+"]
```

```python
def age_grouping(age):
    if age < 18:
        return "<18"
    elif age >= 18 and age < 25:
        return "18-24"
    elif age >=25 and age < 35:
        return "25-34"
    elif age >=35 and age < 45:
        return "35-44"
    elif age >=45 and age < 55:
        return "45-54"
    elif age >=55 and age < 65:
        return "55-64"
    elif age > 65:
        return "65+"
    else:
        return np.nan
    
```

```python
df['age_group'] = df.Age.apply(lambda x: age_grouping(x))
df[['Age', 'age_group']].head(30)
```

```python
# Plot distributions of number of impressions and CTR for each age group
```

```python
df['ctr'] = df.Clicks / df.Impressions 

```

```python
for index, group in df[df.ctr>0].groupby('age_group'):
    #print group.ctr.head()
    n, bins, patches = plt.hist(group.ctr.values, bins=20, normed=1, alpha=0.75)
    plt.xlabel('CTR')
    plt.ylabel('Probability')
    plt.title('Histogram of group: %s' % index)
    plt.show()
```

```python
df[df.ctr>0].groupby('age_group').count()
```

```python
f, axes = plt.subplots(7, 1, sharex=True, figsize=(20, 26))

plt.xlabel('CTR')
for i, (index, group) in enumerate(df[df.ctr>0].groupby('age_group')):
    sns.kdeplot(group.ctr.values, alpha=0.5, shade=True, label=index, ax=axes[i])
    
```

```python
for index, group in df[df.ctr>0].groupby('age_group'):
    #print group.ctr.head()
    plt.boxplot(group.ctr.values)
    plt.xlabel('CTR')
    plt.ylabel('Probability')
    plt.title('Boxplot of group: %s' % index)
    plt.show()
```
